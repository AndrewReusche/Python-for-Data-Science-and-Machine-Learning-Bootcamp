```python
import seaborn as sns 
%matplotlib inline
tips = sns.load_dataset('tips')
tips.head()
```

```python
import numpy as np
```

```python
#bar plot
#sns.barplot(x=,y-,data=, etimator=)
```

```python
sns.barplot(x='sex',y='total_bill',data=tips)
```

```python
sns.barplot(x='sex',y='total_bill',data=tips, estimator='std')
```

```python
sns.barplot(x='sex',y='total_bill',data=tips, estimator=np.std)
```

```python
#countplot
#here the estimator only counts the number of occurances
```

```python
sns.countplot(x='sex', data=tips)
```

```python
#boxplot also known as a box and whisker plot
```

```python
sns.boxplot(x='day',y='total_bill',data=tips)
```

```python
sns.boxplot(x='day',y='total_bill',data=tips, hue='smoker')
```

```python
#violin plot
```

```python
sns.violinplot(x='day',y='total_bill',data=tips)
```

```python
sns.violinplot(x='day',y='total_bill',data=tips,hue='smoker')
```

```python
sns.violinplot(x='day',y='total_bill',data=tips,hue='smoker', split=True)
```

```python
#strip plot
```

```python
sns.stripplot(x='day',y='total_bill', data=tips)
```

```python
sns.stripplot(x='day',y='total_bill', data=tips, jitter=False)
```

```python
sns.stripplot(x='day',y='total_bill', data=tips, jitter=True)
```

```python
sns.stripplot(x='day',y='total_bill', data=tips, jitter=True, hue='sex')
```

```python
sns.stripplot(x='day',y='total_bill', data=tips, jitter=True, hue='sex', dodge=True)
```

```python
#swarm plot
```

```python
sns.swarmplot(x='day',y='total_bill', data=tips)
```

```python
#how to combine swarm plot with violin plot
```

```python
sns.violinplot(x='day',y='total_bill',data=tips)
sns.swarmplot(x='day',y='total_bill', data=tips)
```

```python
#cat plot
```

```python
sns.catplot(x='day',y='total_bill', data=tips, kind ='bar')
```

```python
sns.catplot(x='day',y='total_bill', data=tips, kind ='violin')
```
