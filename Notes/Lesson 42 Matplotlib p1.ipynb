```python
fig = plt.figure()

ax = fig.add_axes([0,0,1,1])

#only show plot between 0 and 1 on the x axes
ax.plot(x,y, color = 'green', linewidth =2, linestyle= '--')
ax.set_xlim(0,1)
```

```python
fig = plt.figure()

ax = fig.add_axes([0,0,1,1])

#only show plot between 0 and 2 on the y axes
ax.plot(x,y, color = 'green', linewidth =2, linestyle= '--')
ax.set_xlim(0,1)
ax.set_ylim(0,2)
```
