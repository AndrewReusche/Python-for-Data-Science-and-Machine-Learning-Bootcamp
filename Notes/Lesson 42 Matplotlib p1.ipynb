```python
#plt.subplot(#of rows,#of colums, name of plot i am reffering to)
```

```python
plt.subplot(1,2,1)
plt.plot(x,y,'r')

plt.subplot(1,2,2)
plt.plot(y,x,'b')

```

```python
plt.subplot(1,2,1)
plt.plot(x,y,'r')
plt.xlabel('X1 Label')
plt.ylabel('Y1 Label')
plt.title('Title1')

plt.subplot(1,2,2)
plt.plot(y,x,'b')
plt.xlabel('X2 Label')
plt.ylabel('Y2 Label')
plt.title('Title2')
```
